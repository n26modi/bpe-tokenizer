```python
# text from https://www.reedbeta.com/blog/programmers-intro-to-unicode/
text = "Ｕｎｉｃｏｄｅ! 🅤🅝🅘🅒🅞🅓🅔‽ 🇺‌🇳‌🇮‌🇨‌🇴‌🇩‌🇪! 😄 The very name strikes fear and awe into the hearts of programmers worldwide. We all know we ought to “support Unicode” in our software (whatever that means—like using wchar_t for all the strings, right?). But Unicode can be abstruse, and diving into the thousand-page Unicode Standard plus its dozens of supplementary annexes, reports, and notes can be more than a little intimidating. I don’t blame programmers for still finding the whole thing mysterious, even 30 years after Unicode’s inception."
tokens = text.encode("utf-8") # raw bytes
tokens = list(map(int, tokens)) # convert to a list of integers in range 0..255 for convenience
print('---')
print(text)
print("length:", len(text))
print('---')
print(tokens)
print("length:", len(tokens)) # num of bytes is different from the num of chars, bc some chars are represented by multiple bytes in UTF-8 (the emoji and other characters at the start)
```

```python
# now find the pair of bytes that occur together often, which is step 1 of BPE algorithm. Then, we'll merge that pair and mint a new token with id of 256. currently tokens go from 0 - 255. 

def get_stats(ids):
    counts = {} #empty dict to sore pair frequencies. Keys are tuples like (101, 32), values are how many times that pair appeared.
    for pair in zip(ids, ids[1:]): # iterate consective elements. zip pairs them up so u get every consecutive pair (ids[0], ids[1]), (ids[1], ids[2]), etc. 
        counts[pair] = counts.get(pair, 0) + 1 #looks up the current count for that pair, defaulting to 0 if it hasn't been seen yet. Then adds 1 and stores it back.
    return counts

stats = get_stats(tokens)

# find the most common pair, stats.items() gives you both the pair and count
top_pair, count = max(stats.items(), key=lambda x: x[1]) # key=lambda x: x[1] tells max to compare by the count (second element), not the pair tuple itself.
print(f"{top_pair} : " , count)
```
